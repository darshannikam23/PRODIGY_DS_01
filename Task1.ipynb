```python
wp_final = wp.drop(['Country Code', 'Indicator Name', 'Indicator Code'], axis = 1) #drop() deletes the whole column
```

```python
chosen_country = 'India'             #choose a country of your choice


country_data = wp_final[wp_final['Country Name'] == chosen_country]   #accessing the Dataset


years = country_data.columns[2:].astype(int)            #extracting years' data from the dataset
population = country_data.iloc[:, 2:].values.flatten()  ##extracting population's data from the dataset

plt.figure(figsize=(5, 3))
plt.bar(years, population, color='cyan')        #plotting size of the bar
plt.xlabel('Year')                              #plotting X axis
plt.ylabel('Population')                        ##plotting Y axis
plt.title(f'Population of {chosen_country} Over the Years')    #plotting title of the bar graph


plt.show()
```

```python
plt.hist(population, bins=20, color='orange', edgecolor='black')    #plotting histogram
plt.show()
```
